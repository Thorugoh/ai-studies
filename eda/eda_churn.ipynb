# Lidando com dados ausentes

* MCAR (Missing completely at random) - O fato de que um certo valor est√° faltando n√£o tem nada a ver com o seu valor hipot√©tico e com os valores de outras vari√°veis.
* MAR (Missing at random) - Faltar dados aleatoriamente significa que a propens√£o para um ponto de dados estar ausente n√£o est√° relacionada a alguns dos dados observados.
* MNAR (Missing not at random) - Duas raz√µes poss√≠veis s√£o que o valor ausente depende do valor hipot√©tico ou o valor ausente depende do valor de alguma outra vari√°vel.

# Formulando hip√≥teses

* Use a intui√ß√£o - Comece com suas suspeitas iniciais com base no conhecimendo do dom√≠nio. Pergunte a si mesmo o que voce espera encontrar nos dados.
* Seja espec√≠fico - Suas hip√≥teses devem ser claras e espec√≠ficas. Evite afirma√ß√µes vagas, como "os dados t√™m alguma tend√™ncia". Em vez disso seja concreto, como "o aumento das vendas est√° relacionado ao lan√ßamento de de um novo produto"
* Testabilidade - Certifique-se de que suas hip√≥teses possam ser testadas com os dados dispon√≠veis. Voc√™ deve ser capaz de encontrar evid√™ncias nos dados que confirmem ou refutem a hip√≥tese.

* Utilize visualiza√ß√µes - A EDA frequentemente envolve gr√°ficos e visualiza√ß√µes. Use os para explorar seus dados e validar suas hip√≥teses. Por exemplo, crie gr√°ficos de dispers√£o para averificar rela√ß√µes entre vari√°veis.
* Seja aberto a surpresas - Esteja preparado para encontrar resultados Inesperados. √Äs vezes as melhores descobertas ocorrem quando as hip√≥teses iniciais s√£o desafiadas.
* Evite hip√≥teses de causa e efeito prematuras - Evite assumir causalidade imediatamente. Por exemplo "A promo√ß√£o de um produto causou um aumento nas vendas". Primeiro explore a correla√ß√£o e, em seguida, avalie a causalidade.
* Refine e Atualize - √Ä medida que voc√™ realiza a EDA e coleta mais informa√ß√µes, refine e atualize suas hip√≥teses como necess√°rio. Esse √© um processo iterativo.
